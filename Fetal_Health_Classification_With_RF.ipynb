# **1. Data Preprocessing**

```python
#Load dataset
import pandas as pd

fetal_health = pd.read_csv('fetal_health.csv')

print('Dataframe: ', fetal_health.shape)
fetal_health.head(10)
```

```python
#Drop duplicates
fetal_health.drop_duplicates(inplace=True)
fetal_health.duplicated().sum()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

#Standarize data
x = fetal_health.drop(['fetal_health'], axis=1) #features
y = fetal_health['fetal_health'] #target class
scaler = StandardScaler()
scaler.fit(x)
standarized = scaler.transform(x)
x = standarized
```

# **2. Model Training & Testing**

```python
#Split Data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

#training & tuning
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(x_train, y_train)

#testing
y_test_predict = model.predict(x_test)
```

#**3. Evaluation**
